```python
from IPython.display import Image

# Specify the path to your PNG image file
image_path = "../Content/PlayerSprites/Shinobi/Idle.png"

# Display the image
Image(filename=image_path)
```

```python
# Extract alpha channel from the image so we can use it as a mask for sprite collision detection
from PIL import Image as PILImage

# Load the image
image = PILImage.open(image_path)

# Extract the alpha channel
alpha = image.getchannel("A")

# Display the alpha channel
alpha
```

```python
# Export the alpha channel as a new image to ./output/{filename}_alpha.png
filename = image_path.split("/")[-1].split(".")[0]
alpha.save(f"./output/{filename}_alpha.png")
```
